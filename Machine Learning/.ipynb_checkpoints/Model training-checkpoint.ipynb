```python
from PIL import Image
import numpy as np
from IPython.display import display

def showImage(image_array):
    display_image = Image.fromarray(image_array)
    display(display_image)

def expand_image(image):
    #image = image.resize([300,300])
    return image

def process_image(image):
    image[0:10, :] = 0
    image[25:48, :] = 0
    return image

from sklearn.model_selection import train_test_split
import os
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv3D, MaxPooling3D, Flatten, Dense


# Directories where the nii files are stored
no_stress = '/home/vedicsharma/workingDirectory/Project2/facesData/train/nostress/'
stress = '/home/vedicsharma/workingDirectory/Project2/facesData/train/stress/'

no_stress_test = '/home/vedicsharma/workingDirectory/Project2/facesData/test/nostress/'
stress_test = '/home/vedicsharma/workingDirectory/Project2/facesData/test/stress/'

# files in directories
no_stress_files = [np.array(expand_image(Image.open(no_stress + f))) for f in os.listdir(no_stress) if f.endswith('.jpg')]  # Only takes files in .nii format
stress_files = [np.array(expand_image(Image.open(stress+ f))) for f in os.listdir(stress) if f.endswith('.jpg')]
#no_stress_files = [os.path.join(no_stress, f) for f in no_stress_files]
#stress_files = [os.path.join(stress, f) for f in stress_files]
no_stress_test_files = [np.array(expand_image(Image.open(no_stress_test + f))) for f in os.listdir(no_stress_test) if f.endswith('.jpg')]  # Only takes files in .nii format
stress_test_files = [np.array(expand_image(Image.open(stress_test+ f))) for f in os.listdir(stress_test) if f.endswith('.jpg')]


# 0 for male, 1 for female
no_stress_labels = [0]*len(no_stress_files)
stress_labels = [1]*len(stress_files)

no_stress_test_labels = [0]*len(no_stress_test_files)
stress_test_labels = [1]*len(stress_test_files)

# X is file paths, Y is labels
X = no_stress_files + stress_files
Y = no_stress_labels + stress_labels

Tx = no_stress_test_files+ stress_test_files
Ty = no_stress_test_labels+stress_test_labels


# Convert to np
X = np.array(X)
Y = np.array(Y)
X = X/255

Tx = np.array(Tx)
Ty = np.array(Ty)
Tx = Tx/255

X_train = [process_image(f) for f in X]
X_train = np.array(X_train)

t_train = [process_image(f) for f in Tx]
t_train = np.array(t_train)





from keras.regularizers import l1, l2, l1_l2
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Dense,
    Conv2D,
    MaxPool2D,
    Flatten,
    Dropout,
    BatchNormalization,
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

reg = reg = l1(0.01)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid', kernel_regularizer=reg)
])
model.summary()


# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])

X_train_, X_val, Y_train, Y_val = train_test_split(
    X_train, Y, test_size=0.2, stratify=Y, random_state=42)

model.fit(X_train_, Y_train, validation_data=(X_val, Y_val), epochs=100, batch_size=128)
score = model.evaluate(
    t_train,
    Ty,
    verbose=0,
)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
